- Crie uma Lista com  os papagaios
# Papagaios
# Feature [1  sim, 0 nao]
# tem bico?
# risco de extinçao
# Fazcocorico

```python
papagaio1 = [1,1,0]
papagaio2 = [1,1,0]
papagaio3 = [1,1,1]
```

```python
# Galinha
# Feature [ 1 sim, 0 Nao]
# Tem bico?
# Risco de extinçao?
# Faz Cocoricó

galinha1 = [1,0,1]
galinha2 = [1,0,1]
galinha3 = [1,1,1]

```

```python
dados = [papagaio1,papagaio2,papagaio3,galinha1,galinha2,galinha3]
```

```python
dados = [papagaio1,papagaio2,papagaio3,galinha1,galinha2,galinha3]
classes = [1,1,1,0,0,0]
```

```python
dados
```

```python
! pip install scikit-learn
```

```python
from sklearn.svm import LinearSVC
modelo = LinearSVC()
```

```python
modelo.fit(dados,classes)
```

prevendo um animal

```python
# features [1 sim, 0 nao]
# Tem Bico?
# Risco de extinçao?
# Faz Cocorico?

# Tem Bico, estaa em extinçao e nao faz cocorico

animal_misterioso = [1,1,0]
previsao = modelo.predict([animal_misterioso])

```

```python
previsao
```

- Previsao de uma Galinha

```python
# Features [1 Sim, 0 Nao]
# Tem Bico?
# Risco de extinçao
# Faz Cociricó

animal_misterioso2 = [1,0,1]
previsao2 = modelo.predict([animal_misterioso2])
```

 - Calculando a Acuracia do modelo
 - Dados e rotulos

```python
# Dados de treino
treino_x = [ papagaio1, papagaio2,papagaio3,galinha1,galinha2,galinha3]
# rotulos(classes) treino y
treino_y = [1,1,1,0,0,0]
```

```python
# Animais misteriosos para teste
misterio1 = [1,1,1]
misterio2 = [1,1,0]
misterio3 = [0,1,1]
```

```python
# previsao dados de teste
teste_x = [misterio1,misterio2,misterio3]
# preendo os teste
previsoes = modelo.predict(teste_x)

```

```python
# rotulos(classes) de teste avaliaçao - provinha
teste_y = [0,1,1]
```

```python
# Importando modulo que ja traz a funçao de acuracia incapsulada
from sklearn.metrics import accuracy_score

# Definindo uma variavel para receber o retorno da funçao acuracia * 100 para porcentagem
taxa_de_acerto = accuracy_score(teste_y,previsoes)* 100

# mostrando a acuracia
print(f"Acuracia:{taxa_de_acerto:.2f}%")
```
